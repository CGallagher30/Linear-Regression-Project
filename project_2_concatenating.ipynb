```python
# Importing Packages
import pandas as pd
```

```python
# List of years
years_int=list(range(2001,2023))

# Convert the years into strings
years=[]
for i in years_int:
    years.append(str(i))
```

```python
main_df=pd.DataFrame()
```

```python
for year in years:
    df1=pd.read_csv('Database/main%s_df.csv'%year)
    main_df=pd.concat([main_df,df1])
```

```python
main_df = main_df.iloc[: , 1:]
```

```python
main_df_names=main_df['Name']
```

```python
main_df=main_df.iloc[: , 1:]
```

```python
main_df=main_df.astype(float)
```

```python
main_df['Name']=main_df_names
```

```python
main_df=main_df.fillna(0)
```

```python
main_df.info()
```

```python
main_df_columns = main_df.columns.tolist()
```

```python
main_df_columns = ['Name','Year','Scoring Average (Actual)','Stroke Differential Field Average','Par 3 Birdie or Better Leaders','Par 4 Birdie or Better Leaders','Par 5 Birdie or Better Leaders','Par 3 Performance','Par 4 Performance','Par 5 Performance','Par Breakers','Birdie Average','Bounce Back','Birdie to Bogey Ratio','Birdie or Better Percentage','Bogey Avoidance','Reverse Bounce Back','Bogey Average','Driving Distance','Driving Accuracy Percentage','Good Drive Percentage','Total Driving','Ball Striking','Greens in Regulation Percentage','GIR Percentage from Fairway','GIR Percentage from Other than Fairway','Scrambling','One-Putt Percentage','1-Putts per Round','3-Putt Avoidance','3-Putts per Round','3+ Putts per Round','Putts Per Round','2-Putts per Round','Putting Average','Overall Putting Average','Birdie or Better Conversion Percentage']
```

```python
main_df=main_df[main_df_columns]
```

```python
main_df.to_csv('csv/main_allyears_df.csv')
```
