```python
# List of statistic code numbers for urls
stat_num=['108','02417','112','113','114','105','156','160','02415','352','02414','02416','02419','101','102','02438','129','158','103','190','199','130','413','398','426','400','401','119','399','104','402','115']
stat_num_spec=['171','172','173']
```

```python
# Scraping year 2021
# Base link = https://www.pgatour.com/stats/stat.108.y2021.eon.t536.html
```

```python
# Importing packages
import time, random
import requests
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
# 2021 Masters Dataframe Creation

# Looping through each stat number

# inital the first url to start scraping
url='https://www.pgatour.com/stats/stat.108.y2021.eon.t536.html'

# initial the first stat
prev_stat=stat_num[0]

main2021_df=pd.DataFrame()

# Start looping
for stat in stat_num:
    url=url.replace(prev_stat,stat)
    prev_stat=stat
    time.sleep(1+2*random.random())
    response=requests.get(url)
    time.sleep(1+2*random.random())
    print(response.status_code)
    html=response.text
    soup=bs(html)
    
    player_rows=soup.find_all(id=re.compile('playerStatsRow'))
    player_list=[row.find(class_='player-name').next.next.next for row in player_rows]
    stat_name=soup.find('h1').text
    print(stat_name)
    stat_value=[row.find(class_='hidden-small hidden-medium').next.next.next.next for row in player_rows]
    row_list=list(zip(player_list,stat_value))
    row_dict_list=[]
    for i in row_list:
        dict={}
        dict['Name']=i[0]
        dict[stat_name]=i[1]
        row_dict_list.append(dict)
    stat_df=pd.DataFrame(row_dict_list)
    prev_df=stat_df
    if main2021_df.empty:
        main2021_df=stat_df
    else:
        main2021_df=pd.merge(main2021_df, stat_df, on = ['Name'], how = 'outer')
    

    
# Looping through each special stat number
# These stats require a slightly different for loop

# inital the first url to start scraping
url='https://www.pgatour.com/stats/stat.171.y2021.eon.t536.html'

# initial the first stat
prev_stat=stat_num_spec[0]


# Start looping
for stat in stat_num_spec:
    url=url.replace(prev_stat,stat)
    prev_stat=stat
    time.sleep(1+2*random.random())
    response=requests.get(url)
    time.sleep(1+2*random.random())
    print(response.status_code)
    html=response.text
    soup=bs(html)
    
    player_rows=soup.find_all(id=re.compile('playerStatsRow'))
    player_list=[row.find(class_='player-name').next.next.next for row in player_rows]
    stat_name=soup.find('h1').text
    print(stat_name)
    stat_value=[row.find(class_='hidden-small hidden-medium').next.next.next.next.next.next.next for row in player_rows]
    row_list=list(zip(player_list,stat_value))
    row_dict_list=[]
    for i in row_list:
        dict={}
        dict['Name']=i[0]
        dict[stat_name]=i[1]
        row_dict_list.append(dict)
    stat_df=pd.DataFrame(row_dict_list)
    prev_df=stat_df
    if main2021_df.empty:
        main2021_df=stat_df
    else:
        main2021_df=pd.merge(main2021_df, stat_df, on = ['Name'], how = 'outer')
    #stat_df.to_csv('csv/%s2021.csv'%stat_name)
main2021_df['Year']='2021'
main2021_df.to_csv('csv/main2021_df.csv')
```
